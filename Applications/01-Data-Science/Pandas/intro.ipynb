```python
import pandas as pd
```

# Series

A series is a one dimensional array of data

```python
print('\nSeries:\n')

pop_list = [724305, 65558, 64674, 89438]

# index is used to give the series a specified index set
series_pop = pd.Series(pop_list, index=['seattle', 'redmond', 'sammamish', 'kirkland'])

print(series_pop)

# Print out series data by index name
print('\nPopulation in Seattle:', series_pop['seattle'])
```

# Dataframe

A dataframe is a two dimensional array of data

```python
print('\nDataframe:\n')

populations = {
  'city': ['seattle', 'redmond', 'sammamish', 'kirkland'],
  'population': [724305, 65558, 64674, 89438]
}

df_pop = pd.DataFrame(populations)

df_pop
```

### Find data at a specific index of the dataframe

```python
print('\nData at index 2:\n')
print(df_pop.loc[[2]]) 
```

### Find data at multiple indexes of the dataframe

```python
print('\nData at index 1 and 3:\n')
print(df_pop.loc[[1, 3]]) 
```

### Give specific index labels to the rows instead of the standard 0,1,2...

```python
annual_pop = {
  'men': [357372, 354461, 354015, 353846],
  'women': [351451, 349454, 349103, 348463]
}

df_annual_pop = pd.DataFrame(annual_pop, index=[2019, 2018, 2017, 2016])

df_annual_pop
```
