# Control Flow

Control flow is how you tell Python **which code to run, when to run it, and how many times**. Without it, every program would just execute line by line from top to bottom — the same way, every time.

## What You'll Learn
1. `if` / `elif` / `else` — conditional branching
2. `match` / `case` — structural pattern matching (Python 3.10+)
3. `for` loops — iterating over sequences
4. `while` loops — looping until a condition is met
5. `break`, `continue`, `pass` — loop control
6. `else` on loops — the underused gem
7. Comprehensions — list, dict, set, generator
8. Exception handling — `try` / `except` / `else` / `finally`
9. Context managers — `with` statements
10. Truthiness & short-circuit evaluation

## 1. `if` / `elif` / `else` — Conditional Branching

The `if` statement is the most fundamental control flow tool. Python uses **indentation** (4 spaces) to define blocks — no curly braces needed.

```python
# Basic if / elif / else
score = 78

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print(f"Score: {score} → Grade: {grade}")
```

```python
# Comparison operators
x, y = 10, 20

print(x == y)   # Equal
print(x != y)   # Not equal
print(x < y)    # Less than
print(x > y)    # Greater than
print(x <= 10)  # Less than or equal
print(x >= 10)  # Greater than or equal

# Python allows chained comparisons — very readable!
age = 25
print(18 <= age < 65)   # True — between 18 and 65

value = 50
print(0 < value < 100)  # True — between 0 and 100
```

```python
# Logical operators: and, or, not
has_ticket = True
is_vip = False
age = 21

if has_ticket and age >= 18:
    print("Welcome in!")

if is_vip or age >= 65:
    print("Special entrance for you.")
else:
    print("Please use the main entrance.")

if not is_vip:
    print("No VIP access.")
```

```python
# Identity and membership operators
x = None
items = [1, 2, 3, 4, 5]

# 'is' checks object identity (use for None, True, False)
if x is None:
    print("x has no value")

if x is not None:
    print("x has a value")

# 'in' checks membership
if 3 in items:
    print("3 is in the list")

if 9 not in items:
    print("9 is not in the list")
```

```python
# Ternary / conditional expression: value_if_true if condition else value_if_false
age = 20
status = "adult" if age >= 18 else "minor"
print(status)

# Practical uses
items = []
message = f"Found {len(items)} items" if items else "No items found"
print(message)

x = -5
absolute = x if x >= 0 else -x
print(f"Absolute value of {x}: {absolute}")
```

## 2. `match` / `case` — Structural Pattern Matching (Python 3.10+)

`match` is Python's modern alternative to long `if/elif` chains. It goes far beyond a simple switch statement — it can match **shapes, types, and structures**.

```python
# Basic value matching — like a switch/case in other languages
def http_status(code):
    match code:
        case 200:
            return "OK"
        case 201:
            return "Created"
        case 301 | 302:         # OR pattern — match either
            return "Redirect"
        case 404:
            return "Not Found"
        case 500:
            return "Internal Server Error"
        case _:                 # wildcard — matches anything (like 'else')
            return f"Unknown status: {code}"

for code in [200, 302, 404, 418]:
    print(f"{code}: {http_status(code)}")
```

```python
# Matching sequences (lists/tuples)
def describe_point(point):
    match point:
        case (0, 0):
            return "Origin"
        case (x, 0):
            return f"On x-axis at {x}"
        case (0, y):
            return f"On y-axis at {y}"
        case (x, y):
            return f"Point at ({x}, {y})"

for pt in [(0, 0), (5, 0), (0, 3), (4, 7)]:
    print(describe_point(pt))
```

```python
# Matching with guards (if conditions)
def classify_number(n):
    match n:
        case 0:
            return "zero"
        case n if n < 0:
            return f"{n} is negative"
        case n if n % 2 == 0:
            return f"{n} is positive and even"
        case n:
            return f"{n} is positive and odd"

for n in [0, -3, 8, 7]:
    print(classify_number(n))
```

```python
# Matching dictionaries (mapping patterns)
def handle_event(event):
    match event:
        case {"type": "click", "button": button}:
            return f"Mouse click: {button} button"
        case {"type": "keypress", "key": key}:
            return f"Key pressed: {key}"
        case {"type": "scroll", "direction": d, "amount": a}:
            return f"Scrolled {d} by {a}px"
        case _:
            return "Unknown event"

events = [
    {"type": "click", "button": "left"},
    {"type": "keypress", "key": "Enter"},
    {"type": "scroll", "direction": "down", "amount": 120},
    {"type": "resize"},
]

for event in events:
    print(handle_event(event))
```

```python
# Matching types with class patterns
def describe(value):
    match value:
        case int(n) if n < 0:
            return f"Negative int: {n}"
        case int(n):
            return f"Positive int: {n}"
        case float(f):
            return f"Float: {f:.2f}"
        case str(s) if len(s) == 0:
            return "Empty string"
        case str(s):
            return f"String: '{s}'"
        case list() as lst:
            return f"List with {len(lst)} items"
        case None:
            return "Nothing (None)"
        case _:
            return f"Something else: {type(value).__name__}"

for val in [-5, 42, 3.14, "", "hello", [1, 2, 3], None]:
    print(describe(val))
```

## 3. `for` Loops — Iterating Over Sequences

Python's `for` loop iterates over any **iterable** — lists, strings, ranges, dicts, files, and more.

```python
# Iterating over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

print()

# Iterating over a string
for char in "Python":
    print(char, end=" ")
print()  # newline
```

```python
# range() — generate a sequence of numbers
# range(stop)
for i in range(5):
    print(i, end=" ")
print()

# range(start, stop)
for i in range(2, 8):
    print(i, end=" ")
print()

# range(start, stop, step)
for i in range(0, 20, 3):
    print(i, end=" ")
print()

# Counting down
for i in range(10, 0, -1):
    print(i, end=" ")
print("🚀")
```

```python
# enumerate() — get index AND value together
fruits = ["apple", "banana", "cherry"]

# ❌ Avoid: manually tracking index
# i = 0
# for fruit in fruits:
#     print(i, fruit)
#     i += 1

# ✅ Pythonic: use enumerate()
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")

print()

# Start counting from a different number
for i, fruit in enumerate(fruits, start=1):
    print(f"{i}. {fruit}")
```

```python
# zip() — iterate over multiple sequences in parallel
names  = ["Alice", "Bob", "Carol"]
scores = [92, 85, 78]
grades = ["A", "B", "C"]

for name, score, grade in zip(names, scores, grades):
    print(f"{name:<8} {score:>3}  ({grade})")

print()

# zip stops at the shortest iterable
# Use itertools.zip_longest to keep going
from itertools import zip_longest
a = [1, 2, 3]
b = ["a", "b"]
for x, y in zip_longest(a, b, fillvalue="?"):
    print(x, y)
```

```python
# Iterating over dictionaries
person = {"name": "Alice", "age": 30, "city": "New York"}

# Keys only (default)
for key in person:
    print(key)
print()

# Values only
for value in person.values():
    print(value)
print()

# Both key and value — most common
for key, value in person.items():
    print(f"{key}: {value}")
```

```python
# Nested loops
rows = ["A", "B", "C"]
cols = [1, 2, 3]

for row in rows:
    for col in cols:
        print(f"{row}{col}", end="  ")
    print()  # newline after each row

print()

# Multiplication table
for i in range(1, 6):
    for j in range(1, 6):
        print(f"{i*j:3}", end="")
    print()
```

## 4. `while` Loops — Loop Until a Condition is False

`while` loops keep running **as long as their condition is True**. Use them when you don't know in advance how many iterations you need.

```python
# Basic while loop
count = 0
while count < 5:
    print(f"count = {count}")
    count += 1   # ⚠️ Always update the condition variable or risk an infinite loop!

print("Done.")
```

```python
# Countdown
n = 5
while n > 0:
    print(n, end=" ")
    n -= 1
print("Liftoff! 🚀")
```

```python
# Simulating user input with a sentinel value
# (We'll use a list to simulate input instead of actual input())
simulated_inputs = ["hello", "world", "quit"]
idx = 0

while True:   # ← infinite loop — must break out manually
    user_input = simulated_inputs[idx]
    idx += 1
    print(f"  User typed: {user_input}")

    if user_input.lower() == "quit":
        print("Goodbye!")
        break
    print(f"  Echo: {user_input.upper()}")
```

```python
# while with a flag variable — clean way to manage complex exit conditions
running = True
attempts = 0
max_attempts = 3
secret = 42
guesses = [10, 42, 99]  # simulated guesses

while running:
    guess = guesses[attempts]
    attempts += 1
    print(f"Guess #{attempts}: {guess}")

    if guess == secret:
        print("Correct! 🎉")
        running = False
    elif attempts >= max_attempts:
        print(f"Out of attempts. The answer was {secret}.")
        running = False
    else:
        hint = "too low" if guess < secret else "too high"
        print(f"Wrong — {hint}. Try again.")
```

## 5. `break`, `continue`, `pass` — Loop Control Statements

```python
# break — exit the loop immediately
print("Finding first even number > 10:")
for n in range(11, 30):
    if n % 2 == 0:
        print(f"Found: {n}")
        break   # stop as soon as we find it

print()

# break in nested loops only breaks the INNER loop
print("Nested break (inner only):")
for i in range(3):
    for j in range(3):
        if j == 1:
            break   # only breaks the j-loop
        print(f"  ({i},{j})", end="")
    print()
```

```python
# continue — skip the rest of this iteration and go to the next
print("Odd numbers only:")
for n in range(10):
    if n % 2 == 0:
        continue   # skip even numbers
    print(n, end=" ")
print()

print()

# Practical use: skip invalid/unwanted data
data = [10, None, 25, None, 30, 0, 45]
total = 0
for value in data:
    if value is None or value == 0:
        continue
    total += value
print(f"Sum (skipping None and 0): {total}")
```

```python
# pass — do absolutely nothing (a no-op placeholder)
# Useful when Python requires a statement but you have nothing to write yet

# Placeholder for future code
for i in range(5):
    pass   # TODO: implement later

# Empty if branch
x = 10
if x > 5:
    pass   # handle later
else:
    print("x is small")

# Empty class or function body
class MyFutureClass:
    pass

def not_implemented_yet():
    pass

print("pass lets you write empty blocks without syntax errors.")
```

## 6. `else` on Loops — The Underused Gem

Python's loops have an optional `else` clause. It runs **only if the loop completed without hitting a `break`**. This is surprisingly useful.

```python
# for...else: the else runs only if break was NOT triggered
def find_prime(numbers):
    for n in numbers:
        for divisor in range(2, n):
            if n % divisor == 0:
                break           # n is not prime
        else:
            # This only runs if the inner loop finished WITHOUT break
            print(f"{n} is prime!")
            return n
    print("No prime found")
    return None

find_prime([4, 6, 8, 9, 11])
```

```python
# Another classic use: searching for an item
def find_user(users, target):
    for user in users:
        if user["name"] == target:
            print(f"Found: {user}")
            break
    else:
        # Only runs if we never broke out — meaning target was NOT found
        print(f"User '{target}' not found.")

users = [
    {"name": "Alice", "role": "admin"},
    {"name": "Bob",   "role": "user"},
]

find_user(users, "Alice")
find_user(users, "Carol")
```

```python
# while...else works the same way
attempts = 0
max_attempts = 3
success = False
results = [False, False, True]  # simulated attempt results

while attempts < max_attempts:
    result = results[attempts]
    attempts += 1
    print(f"Attempt {attempts}: {'success' if result else 'failed'}")
    if result:
        print("Operation succeeded!")
        break
else:
    # Only runs if we exhausted all attempts (no break)
    print("All attempts failed.")
```

## 7. Comprehensions — Concise, Readable Collection Building

Comprehensions are a Pythonic way to build collections in a single, readable line.

```python
# List comprehension: [expression for item in iterable if condition]

# ❌ Traditional loop
squares = []
for n in range(10):
    squares.append(n ** 2)

# ✅ List comprehension
squares = [n ** 2 for n in range(10)]
print(squares)

# With a filter condition
even_squares = [n ** 2 for n in range(10) if n % 2 == 0]
print(even_squares)

# Transform strings
words = ["hello", "world", "python"]
upper_words = [w.upper() for w in words]
print(upper_words)

# Filter and transform at once
long_words = [w.capitalize() for w in words if len(w) > 4]
print(long_words)
```

```python
# Nested list comprehension — flatten a 2D list
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

flat = [num for row in matrix for num in row]
print(flat)

# Create a matrix using nested comprehension
grid = [[i * j for j in range(1, 4)] for i in range(1, 4)]
for row in grid:
    print(row)
```

```python
# Dict comprehension: {key: value for item in iterable}
words = ["apple", "banana", "cherry"]

# Map word → its length
word_lengths = {word: len(word) for word in words}
print(word_lengths)

# Invert a dictionary
original = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in original.items()}
print(inverted)

# Filter a dict — keep only pairs where value > 1
filtered = {k: v for k, v in original.items() if v > 1}
print(filtered)
```

```python
# Set comprehension: {expression for item in iterable}
numbers = [1, 2, 2, 3, 4, 4, 4, 5]

unique_squares = {n ** 2 for n in numbers}
print(unique_squares)   # no duplicates, unordered

# Get unique first letters from a list of words
words = ["apple", "avocado", "banana", "blueberry", "cherry"]
first_letters = {w[0] for w in words}
print(first_letters)
```

```python
# Generator expressions: (expression for item in iterable)
# Like list comprehensions but LAZY — values computed one at a time
# Use when you only need to iterate once and want to save memory

# List comprehension — builds the whole list in memory
big_list = [n ** 2 for n in range(1_000_000)]

# Generator expression — computes values on demand
big_gen = (n ** 2 for n in range(1_000_000))

import sys
print(f"List size:      {sys.getsizeof(big_list):,} bytes")
print(f"Generator size: {sys.getsizeof(big_gen):,} bytes")

# Generators work great with sum(), max(), any(), all()
total = sum(n ** 2 for n in range(101))   # no [] needed inside sum()
print(f"Sum of squares 0–100: {total}")

has_negative = any(n < 0 for n in [1, 2, -3, 4])
print(f"Any negatives? {has_negative}")

all_positive = all(n > 0 for n in [1, 2, 3, 4])
print(f"All positive?  {all_positive}")
```

## 8. Exception Handling — `try` / `except` / `else` / `finally`

Exceptions are how Python signals that something went wrong. Handling them gracefully makes your programs robust.

```python
# Basic try/except
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero!")

print()

# Catching the exception object for details
try:
    number = int("not a number")
except ValueError as e:
    print(f"ValueError caught: {e}")
```

```python
# Multiple except clauses — handle different errors differently
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print("Error: division by zero")
        return None
    except TypeError as e:
        print(f"Error: wrong type — {e}")
        return None

print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide(10, "x"))

print()

# Catch multiple exception types in one clause
def parse_index(data, index):
    try:
        return int(data[index])
    except (IndexError, ValueError) as e:
        print(f"Could not parse: {e}")
        return None

print(parse_index(["10", "20", "abc"], 2))
print(parse_index(["10", "20"], 9))
```

```python
# try / except / else / finally — the full form
#
#   try      — code that might fail
#   except   — what to do if it fails
#   else     — runs only if NO exception occurred
#   finally  — ALWAYS runs, exception or not (cleanup code)

def read_number(s):
    try:
        value = int(s)
    except ValueError:
        print(f"  ✗ '{s}' is not a valid number")
    else:
        # Only runs if int(s) succeeded
        print(f"  ✓ Parsed successfully: {value}")
    finally:
        # Always runs — great for releasing resources
        print(f"  (finished processing '{s}')")

read_number("42")
print()
read_number("hello")
```

```python
# Raising exceptions manually with raise
def set_age(age):
    if not isinstance(age, int):
        raise TypeError(f"age must be an int, got {type(age).__name__}")
    if age < 0 or age > 150:
        raise ValueError(f"age must be between 0 and 150, got {age}")
    return age

print(set_age(25))

try:
    set_age(-5)
except ValueError as e:
    print(f"ValueError: {e}")

try:
    set_age("old")
except TypeError as e:
    print(f"TypeError: {e}")
```

```python
# Creating custom exceptions
class InsufficientFundsError(Exception):
    """Raised when a withdrawal exceeds the account balance."""
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        super().__init__(f"Cannot withdraw ${amount:.2f}: balance is ${balance:.2f}")

def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientFundsError(balance, amount)
    return balance - amount

try:
    new_balance = withdraw(50.00, 75.00)
except InsufficientFundsError as e:
    print(f"Transaction failed: {e}")
    print(f"  Balance: ${e.balance:.2f}")
    print(f"  Attempted: ${e.amount:.2f}")
```

```python
# Common built-in exceptions and when you'll see them
examples = [
    ("ValueError",     lambda: int("abc")),
    ("TypeError",      lambda: "2" + 2),
    ("IndexError",     lambda: [1, 2, 3][10]),
    ("KeyError",       lambda: {"a": 1}["b"]),
    ("AttributeError", lambda: (42).upper()),
    ("ZeroDivisionError", lambda: 1 / 0),
    ("NameError",      lambda: eval("undefined_variable")),
]

for name, fn in examples:
    try:
        fn()
    except Exception as e:
        print(f"{name:22}: {e}")
```

## 9. Context Managers — `with` Statements

The `with` statement provides a clean way to manage **setup and teardown** around a block of code — even if errors occur. File handling is the most common use case.

```python
import tempfile, os

# File I/O — the with statement guarantees the file is closed
# even if an exception is raised inside the block

# Write a file
tmp_path = tempfile.mktemp(suffix=".txt")
with open(tmp_path, 'w', encoding='utf-8') as f:
    f.write("Line 1\n")
    f.write("Line 2\n")
    f.write("Line 3\n")
print(f"Wrote to {tmp_path}")

# Read it back
with open(tmp_path, 'r', encoding='utf-8') as f:
    content = f.read()
print(content)

# Read line by line (memory-efficient for large files)
with open(tmp_path, 'r') as f:
    for i, line in enumerate(f, start=1):
        print(f"  Line {i}: {line.strip()}")

os.unlink(tmp_path)  # clean up temp file
```

```python
# Multiple context managers in one with statement
import tempfile

src = tempfile.mktemp()
dst = tempfile.mktemp()

with open(src, 'w') as f:
    f.write("source content\n")

with open(src, 'r') as infile, open(dst, 'w') as outfile:
    for line in infile:
        outfile.write(line.upper())

with open(dst) as f:
    print(f.read())

os.unlink(src)
os.unlink(dst)
```

```python
# Creating your own context manager with contextlib
from contextlib import contextmanager
import time

@contextmanager
def timer(label):
    """Measure how long a block of code takes."""
    start = time.perf_counter()
    try:
        yield          # ← code inside the 'with' block runs here
    finally:
        elapsed = time.perf_counter() - start
        print(f"[{label}] took {elapsed*1000:.2f}ms")

with timer("list comprehension"):
    result = [n ** 2 for n in range(100_000)]

with timer("join 10k strings"):
    s = "".join(str(i) for i in range(10_000))
```

```python
# suppress — silently ignore specific exceptions
from contextlib import suppress

data = {"name": "Alice"}

# Without suppress — need try/except
try:
    del data["age"]
except KeyError:
    pass

# With suppress — cleaner
with suppress(KeyError):
    del data["age"]    # silently ignored if key doesn't exist

print(data)
```

## 10. Truthiness & Short-Circuit Evaluation

Understanding how Python evaluates truth is key to writing clean, idiomatic code.

```python
# Falsy values — everything else is truthy
falsy_values = [False, None, 0, 0.0, 0j, "", [], (), {}, set()]

print("Falsy values:")
for v in falsy_values:
    print(f"  bool({v!r:12}) = {bool(v)}")

print()
print("Some truthy values:")
for v in [True, 1, -1, "0", [0], {"a": 0}]:
    print(f"  bool({v!r:12}) = {bool(v)}")
```

```python
# Using truthiness in conditions — idiomatic Python
items = []

# ❌ Verbose
if len(items) == 0:
    print("List is empty")

# ✅ Pythonic
if not items:
    print("List is empty")

name = "Alice"
if name:  # checks if name is not empty string or None
    print(f"Hello, {name}!")
```

```python
# Short-circuit evaluation
# 'and' returns the FIRST falsy value, or the LAST value if all truthy
# 'or'  returns the FIRST truthy value, or the LAST value if all falsy

print(0 and 10)       # 0  (first falsy)
print(5 and 10)       # 10 (all truthy — returns last)
print(5 and 0)        # 0  (first falsy after 5)

print(0 or 10)        # 10 (first truthy)
print(False or "")    # '' (all falsy — returns last)
print(False or None or "default")  # 'default'
```

```python
# Practical patterns using short-circuit evaluation

# Default value pattern (common before walrus operator)
user_input = ""       # simulating empty input
name = user_input or "Anonymous"
print(f"Name: {name}")

# Safe attribute/key access
config = None
timeout = config and config.get("timeout")  # won't crash if config is None
print(f"Timeout: {timeout}")

# The walrus operator := (Python 3.8+)
# Assign AND test in one expression
data = [3, 7, 1, 9, 4, 2, 8]

# Find first number > 5 and use it immediately
if found := next((n for n in data if n > 5), None):
    print(f"First number > 5: {found}")

# Walrus in while — process chunks until empty
numbers = iter([10, 20, 30])
while chunk := next(numbers, None):
    print(f"  Processing: {chunk}")
```

## 11. Quick Reference

| Construct | Use When |
|---|---|
| `if / elif / else` | Branch based on conditions |
| `match / case` | Match a value's structure (Python 3.10+) |
| `for` | Iterate a known sequence or iterable |
| `while` | Loop until a condition changes |
| `break` | Exit the current loop immediately |
| `continue` | Skip to the next iteration |
| `pass` | Placeholder — do nothing |
| `for/while...else` | Run cleanup code only if loop didn't `break` |
| `[x for x in ...]` | Build a list concisely |
| `{k:v for ...}` | Build a dict concisely |
| `(x for x in ...)` | Lazy generator — memory-efficient iteration |
| `try/except/else/finally` | Handle errors gracefully |
| `with` | Guaranteed setup/teardown (files, locks, timers) |
| `:=` (walrus) | Assign and test in one expression (Python 3.8+) |

```python
# 🏆 Practice: Mini text adventure using control flow

def play_game():
    print("=" * 40)
    print("     🗺️  THE CONTROL FLOW DUNGEON")
    print("=" * 40)

    rooms = [
        {"name": "Entrance Hall",  "item": "torch",  "monster": None},
        {"name": "Dark Corridor",  "item": None,     "monster": "goblin"},
        {"name": "Treasure Room",  "item": "gold",   "monster": "dragon"},
    ]

    inventory = []
    player_hp = 10

    for room in rooms:
        print(f"\n📍 You enter: {room['name']}")

        # Pick up item
        if item := room["item"]:
            inventory.append(item)
            print(f"  Found a {item}! Added to inventory.")

        # Fight monster
        if monster := room["monster"]:
            print(f"  ⚔️  A {monster} appears!")
            if "torch" in inventory:
                print(f"  You use the torch — the {monster} flees!")
            else:
                damage = 3
                player_hp -= damage
                print(f"  No weapon! You take {damage} damage. HP: {player_hp}")

            if player_hp <= 0:
                print("  You have perished. Game over. 💀")
                break
    else:
        # Loop completed without break — player survived!
        print(f"\n🏆 You escaped the dungeon!")
        print(f"   HP remaining: {player_hp}")
        print(f"   Inventory: {inventory}")

play_game()
```
