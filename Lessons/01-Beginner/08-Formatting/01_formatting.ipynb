## 2. PEP 8 — The Official Style Guide

**PEP 8** (Python Enhancement Proposal 8) is Python's official style guide, authored by Guido van Rossum. It covers naming, whitespace, imports, comments, and more. It's the baseline that all other tools and team style guides build on.

> 📖 Full guide: https://peps.python.org/pep-0008/

```python
# PEP 8 is a guide, not a law — but deviating needs good reason.
# Key principle: "A Foolish Consistency is the Hobgoblin of Little Minds"
# (quoting Emerson) — meaning: apply rules with judgement, not blindly.
#
# When to break a PEP 8 rule:
#   1. Following it would make code LESS readable
#   2. Surrounding code already breaks it (consistency within a file)
#   3. The code predates PEP 8 and changing it is too disruptive

print("PEP 8: a guide for writing readable Python.")
```

## 3. Indentation & Whitespace

```python
# ── Rule: Use 4 spaces per indentation level ─────────────────────────────
# NEVER mix tabs and spaces — Python 3 raises a TabError if you do.
# Most editors convert Tab key to 4 spaces automatically.

# ❌ 2 spaces
def bad_indent():
  x = 1
  if x:
    return x

# ✅ 4 spaces
def good_indent():
    x = 1
    if x:
        return x

print("4 spaces per indent level.")
```

```python
# ── Continuation indentation ──────────────────────────────────────────────
# When a statement spans multiple lines, align with the opening delimiter
# OR add 4 extra spaces to distinguish it from the body.

# ✅ Aligned with opening delimiter
result = some_function(arg_one, arg_two,
                        arg_three, arg_four) if False else None

# ✅ 4-space hanging indent (closing bracket on its own line)
result = some_function(
    arg_one,
    arg_two,
    arg_three,
    arg_four,          # trailing comma is fine and helps diffs
) if False else None

# ✅ Multi-line if condition — add extra indent to distinguish from body
long_variable_name = True
another_long_name = True

if (long_variable_name
        and another_long_name):   # extra 4 spaces vs the body below
    print("Condition met")

# ✅ Closing bracket options — both are acceptable
my_list = [
    1, 2, 3,
    4, 5, 6,
]              # closing bracket lines up with first non-whitespace char

my_dict = {
    "key": "value",
    }          # closing bracket lines up with the opening line — also OK
```

```python
# ── Spaces around operators ───────────────────────────────────────────────

# ❌
x=1
y = x+1
z = x*2+y*3

# ✅ Single space around assignment and binary operators
x = 1
y = x + 1
z = x * 2 + y * 3

# ✅ Exception: no spaces around = in keyword arguments
def connect(host, port=5432, timeout=30):
    pass

connect('localhost', port=5433, timeout=10)   # ← no spaces around =

# ✅ Exception: tighter binding in complex expressions
# When mixing operators of different precedence, use spacing to
# visually group the higher-precedence operations:
hypot = x*x + y*y          # multiplication binds tighter than addition
result = (x+1) * (y-1)     # spaces within groups

print("Operator spacing demonstrated.")
```

```python
# ── Spaces around colons, commas, and brackets ────────────────────────────

# ❌ Spaces before colons/commas
d = {1 : 2 , 3 : 4}
t = (1 , 2 , 3)

# ✅ No space before colon/comma, one space after
d = {1: 2, 3: 4}
t = (1, 2, 3)

# ❌ Spaces inside brackets
s = [ 1, 2, 3 ]
f( x, y )
d[ 'key' ]

# ✅ No spaces inside brackets
s = [1, 2, 3]
# f(x, y)
# d['key']

# ❌ Space before function call parenthesis
# print ('hello')

# ✅ No space before (
print('hello')

# ❌ Space before slice colon
s = [1, 2, 3, 4, 5]
_ = s[1 : 3]
_ = s[1 : 3 : 2]

# ✅ Treat slice colon like a binary operator — consistent spacing
_ = s[1:3]
_ = s[1:3:2]
_ = s[::2]
_ = s[1 : 3]   # OK when emphasising the slice bounds

print("Spacing around punctuation demonstrated.")
```

## 4. Line Length & Line Continuation

```python
# ── PEP 8 rule: limit lines to 79 characters ─────────────────────────────
# In practice, most modern projects use 88 (Black's default) or 100.
# Docstrings and comments: limit to 72 characters per PEP 8,
# though many projects allow 79 or 88 here too.
#
# The real goal: avoid horizontal scrolling. Pick one limit, be consistent.

# Visual ruler at 79 and 88 characters:
print("1234567890" * 7 + "123456789")   # 79 chars
print("1234567890" * 8 + "12345678")    # 88 chars
```

```python
# ── Preferred: implicit continuation inside brackets ─────────────────────
# Python allows you to split any expression inside (), [], {} across lines.
# This is the preferred method — no backslash needed.

# Long function call
result = dict(
    name="Alice",
    age=30,
    city="New York",
    department="Engineering",
)

# Long list / dict literal
colours = [
    "red", "orange", "yellow",
    "green", "blue", "indigo", "violet",
]

# Long boolean condition — wrap in ()
is_valid = (
    len(colours) > 0
    and all(isinstance(c, str) for c in colours)
    and all(len(c) > 0 for c in colours)
)

# Long string
message = (
    "This is a very long message that would exceed the line limit "
    "if written on a single line. Adjacent string literals are "
    "automatically concatenated by Python."
)

print(is_valid)
print(message[:60], "...")
```

```python
# ── Backslash continuation — use sparingly ────────────────────────────────
# The backslash method works but is fragile (trailing space breaks it)
# and less readable. Prefer bracket-based continuation.

# ❌ Backslash continuation
total = 1 + \
        2 + \
        3

# ✅ Bracket-based continuation
total = (
    1
    + 2
    + 3
)

# Exception: with statements before Python 3.10 (no parentheses)
# with open('a') as f1, \
#      open('b') as f2:
#     ...

# Python 3.10+ parenthesised with
# with (
#     open('a') as f1,
#     open('b') as f2,
# ):
#     ...

print(total)
```

```python
# ── Where to break a binary operator line? ───────────────────────────────
# Old PEP 8 (before 2015): break AFTER the operator
# New PEP 8 (W504): break BEFORE the operator — matches math notation

# ❌ Old style — break after operator (harder to scan)
income = (
    gross_wages +
    taxable_interest +
    dividends -
    qualified_dividends
) if False else 0

# ✅ New style — break before operator (operator is at the start of each line)
income = (
    gross_wages
    + taxable_interest
    + dividends
    - qualified_dividends
) if False else 0

print("Line break before operator is now preferred.")
```

## 5. Blank Lines

```python
# ── Blank line rules ──────────────────────────────────────────────────────
#
#  2 blank lines  →  before and after top-level definitions
#                    (functions, classes)
#  1 blank line   →  between methods inside a class
#  1 blank line   →  sparingly inside functions, to separate logical steps
#  0 blank lines  →  between related one-liners

# ── Module level ─────────────────────────────────────────────────────────
CONSTANT_ONE = 1
CONSTANT_TWO = 2
                                            # ← 2 blank lines

def top_level_function():
    """A top-level function."""
    pass
                                            # ← 2 blank lines

def another_top_level():
    """Another top-level function."""

    # ← 1 blank line inside a function to separate logical sections
    step_one = 1
    step_two = 2

    result = step_one + step_two
    return result
                                            # ← 2 blank lines

class MyClass:
    """A class."""

    CLASS_VAR = "shared"                    # ← 1 blank line after class docstring

    def __init__(self):
        self.value = 0
                                            # ← 1 blank line between methods
    def method_one(self):
        """First method."""
        return self.value
                                            # ← 1 blank line between methods
    def method_two(self):
        """Second method."""
        return self.value * 2


print("Blank line rules demonstrated.")
```

## 7. Naming Conventions

```python
# ── Naming convention summary ─────────────────────────────────────────────

naming = """
╔══════════════════════╦══════════════════════╦══════════════════════════╗
║ Identifier           ║ Convention           ║ Examples                 ║
╠══════════════════════╬══════════════════════╬══════════════════════════╣
║ Variable             ║ snake_case           ║ user_name, total_count   ║
║ Function / method    ║ snake_case           ║ get_user(), send_email() ║
║ Constant (module)    ║ UPPER_SNAKE_CASE     ║ MAX_RETRIES, PI, VERSION ║
║ Class                ║ PascalCase           ║ BankAccount, HttpClient  ║
║ Exception            ║ PascalCase + Error   ║ ValidationError, TimeoutError ║
║ Type alias           ║ PascalCase           ║ UserId, JsonData         ║
║ Module / package     ║ lowercase (short)    ║ utils, mypackage         ║
║ Private (convention) ║ _single_leading_     ║ _internal, _cache        ║
║ Name mangling        ║ __double_leading_    ║ __private_attr           ║
║ Dunder / magic       ║ __double_both__      ║ __init__, __str__        ║
╚══════════════════════╩══════════════════════╩══════════════════════════╝
"""
print(naming)
```

```python
# ── Private vs public naming ──────────────────────────────────────────────

class MyClass:
    # Public — part of the class's API
    public_attr = "visible"

    # Private by convention — don't use from outside (but Python doesn't enforce)
    _internal_cache = {}
    _helper_method_result = None

    # Name-mangled — Python renames to _MyClass__attr to avoid subclass clashes
    __private_attr = "truly private"

    # Dunder — special Python-defined meaning, never invent new ones
    def __init__(self):
        self.value = 0

    def public_method(self):
        return self._compute()   # can call private from within

    def _compute(self):          # private helper
        return self.value * 2


obj = MyClass()
print(obj.public_attr)          # ✅ OK
print(obj._internal_cache)      # works, but signals "don't do this"
# print(obj.__private_attr)     # ← AttributeError! Name-mangled
print(obj._MyClass__private_attr)   # works if you really need it

# Trailing underscore — avoid conflicts with Python keywords
class_ = "MyClass"              # avoids shadowing 'class'
type_  = "int"                  # avoids shadowing 'type'
id_    = 42                     # avoids shadowing built-in 'id'
```

## 8. Expressions & Statements

```python
# ── One statement per line ────────────────────────────────────────────────

# ❌ Multiple statements on one line
x = 1; y = 2; z = 3
if x > 0: print(x)

# ✅ One statement per line
x = 1
y = 2
z = 3
if x > 0:
    print(x)

# ✅ Exception: related one-liners can be compact
# (but only when they make the code more readable)
NORTH = 0; SOUTH = 1; EAST = 2; WEST = 3   # related constants — debatable
```

```python
# ── Boolean comparisons ───────────────────────────────────────────────────

items = [1, 2, 3]
value = None
flag  = True

# ❌ Comparing to True/False/None with ==
if flag == True:
    pass
if value == None:
    pass
if len(items) > 0:
    pass

# ✅ Use truthiness and 'is'
if flag:                   # truthy check
    pass
if not flag:
    pass
if value is None:          # identity check for None
    pass
if value is not None:
    pass
if items:                  # non-empty container is truthy
    pass
if not items:              # empty container is falsy
    pass

# ❌ Redundant comparisons
if (flag == True) == True:
    pass

# ✅
if flag:
    pass

print("Use truthiness, not == True/False/None.")
```

```python
# ── Return statements ─────────────────────────────────────────────────────

# ❌ Inconsistent return — sometimes returns a value, sometimes falls off
def get_value_bad(d, key):
    if key in d:
        return d[key]   # returns a value
    # falls off end — implicitly returns None

# ✅ Explicit return on all paths
def get_value_good(d, key):
    if key in d:
        return d[key]
    return None         # explicit None return

# ❌ Bare return when the function always returns None anyway
def setup_bad():
    x = 1
    return   # pointless unless there's conditional logic

# ✅ Either use bare return for early exit, or omit it
def setup_good(flag):
    if not flag:
        return          # early exit — bare return is appropriate here
    # ... continue setup ...

print("Return statement consistency.")
```

```python
# ── Exception handling ────────────────────────────────────────────────────

# ❌ Bare except — catches EVERYTHING including KeyboardInterrupt, SystemExit
try:
    pass
except:
    pass   # never do this

# ❌ Overly broad except
try:
    pass
except Exception:
    pass   # almost as bad

# ✅ Specific exception types
try:
    result = int("abc")
except ValueError as e:
    print(f"Conversion failed: {e}")

# ✅ Multiple specific types
try:
    data = {}
    _ = data["key"]
except (KeyError, TypeError) as e:
    print(f"Lookup error: {e}")

# ❌ Silent exception swallowing
try:
    result = 1 / 0
except ZeroDivisionError:
    pass   # silently ignored — makes debugging very hard

# ✅ At minimum, log the error
import logging
try:
    result = 1 / 0
except ZeroDivisionError:
    logging.warning("Division by zero encountered")
    result = 0
```

## 10. Type Annotation Style

```python
# ── Type hint formatting ──────────────────────────────────────────────────
# PEP 8 section: add a space after the colon, space around -> arrow.

from typing import Optional, Union

# ❌
def bad(x:int, y:str='hello') ->bool:
    return True

# ✅
def good(x: int, y: str = 'hello') -> bool:
    return True

# ✅ Variable annotations
count: int = 0
name: str
items: list[str] = []
mapping: dict[str, int] = {}

# ✅ Modern union syntax (Python 3.10+)
def find(value: int | None) -> str | None:
    return str(value) if value is not None else None

# ✅ Pre-3.10 union syntax
def find_old(value: Optional[int]) -> Optional[str]:
    return str(value) if value is not None else None

print(good(5))
```

```python
# ── Complex type hint formatting ──────────────────────────────────────────
from typing import Callable

# ✅ Long signatures — break like any other long line
def process_records(
    records: list[dict[str, str]],
    transform: Callable[[dict[str, str]], dict[str, str]],
    validate: Callable[[dict[str, str]], bool],
    output_path: str | None = None,
) -> list[dict[str, str]]:
    """Process, transform, and validate a list of records."""
    return [transform(r) for r in records if validate(r)]


# ✅ TypeAlias for repeated complex types (Python 3.12 uses 'type' keyword)
from typing import TypeAlias

Record: TypeAlias = dict[str, str]
RecordList: TypeAlias = list[Record]
Transformer: TypeAlias = Callable[[Record], Record]

def process_v2(
    records: RecordList,
    transform: Transformer,
) -> RecordList:
    """Cleaner signature using type aliases."""
    return [transform(r) for r in records]


print("Type alias makes complex signatures readable.")
```

## 11. Common PEP 8 Violations — Before & After

```python
# ── Before: a messy but working function ─────────────────────────────────

import json,os
from math import *

def ProcessUserData(userData,outputDir,verbose=False):
    """process user data"""
    Results=[]
    for u in userData:
        if u['age']>18 and u['active']==True:
            Score=sqrt(u['score'])
            if Score>5.0 :
                Results.append({'id':u['id'],'score':round(Score,2),'status':'pass'})
            else :
                Results.append({'id':u['id'],'score':round(Score,2),'status':'fail'})
    if verbose==True:
        print( 'Processed', len( Results ), 'users' )
    outPath=os.path.join(outputDir,'results.json')
    with open(outPath,'w') as f: json.dump(Results,f)
    return Results

print("Messy version defined.")
```

```python
# ── After: PEP 8 compliant ────────────────────────────────────────────────

import json
import math
import os
from typing import TypedDict


class UserRecord(TypedDict):
    """A single raw user record."""
    id: int
    age: int
    active: bool
    score: float


class ProcessedRecord(TypedDict):
    """A processed user record with computed score."""
    id: int
    score: float
    status: str


def process_user_data(
    user_data: list[UserRecord],
    output_dir: str,
    verbose: bool = False,
) -> list[ProcessedRecord]:
    """
    Process active adult users and write results to a JSON file.

    Filters to active users aged 18+, applies a square-root score
    normalisation, and classifies each user as 'pass' or 'fail'.

    Args:
        user_data: List of raw user records.
        output_dir: Directory where results.json will be written.
        verbose: Print a summary to stdout if True. Defaults to False.

    Returns:
        List of processed records for qualifying users.
    """
    results = []

    for user in user_data:
        if user["age"] <= 18 or not user["active"]:
            continue

        normalised_score = math.sqrt(user["score"])
        status = "pass" if normalised_score > 5.0 else "fail"

        results.append({
            "id":     user["id"],
            "score":  round(normalised_score, 2),
            "status": status,
        })

    if verbose:
        print(f"Processed {len(results)} users.")

    output_path = os.path.join(output_dir, "results.json")
    with open(output_path, "w", encoding="utf-8") as f:
        json.dump(results, f, indent=2)

    return results


print("Clean version defined.")
```

## 12. Automated Formatters

Manual enforcement of style is exhausting and inconsistent. Modern Python development uses **auto-formatters** that rewrite your code to match a style — no debates, no manual fixes.

```python
# ── black — the uncompromising formatter ──────────────────────────────────
# Black reformats your entire codebase to a consistent style.
# It has almost no configuration by design — that's the point.
# Line length: 88 characters (default). Double quotes.

black_example = '''
# BEFORE black:
def foo( x,y,z ):
    return {"a":x,"b":y,"c":z}

x = [1,2,3,   4,5]
long_variable = some_module.some_function(argument_one,argument_two,argument_three,argument_four)

# AFTER black:
def foo(x, y, z):
    return {"a": x, "b": y, "c": z}


x = [1, 2, 3, 4, 5]
long_variable = some_module.some_function(
    argument_one, argument_two, argument_three, argument_four
)
'''
print(black_example)
```

```python
# ── black usage ───────────────────────────────────────────────────────────
black_usage = """
# Install
pip install black

# Format a single file
black myfile.py

# Format an entire project
black .

# Check without writing (exit code 1 if changes needed)
black --check .

# Show diff without writing
black --diff myfile.py

# Format only specific lines
black --line-ranges 10-20 myfile.py

# Configuration (pyproject.toml):
[tool.black]
line-length = 88
target-version = ['py310', 'py311', 'py312']
skip-string-normalization = false   # keep single quotes if true
"""
print(black_usage)
```

```python
# ── ruff format — the fast modern alternative to black ───────────────────
# ruff is a Rust-based tool that is 10-100x faster than black.
# Its formatter is black-compatible by design.
# ruff also replaces flake8, isort, pyupgrade, and more (see Linters section).

ruff_format_usage = """
# Install
pip install ruff

# Format (equivalent to black .)
ruff format .

# Check without writing
ruff format --check .

# Configuration (pyproject.toml):
[tool.ruff.format]
quote-style = "double"    # or "single"
indent-style = "space"    # or "tab"
line-ending = "auto"      # or "lf", "crlf"
"""
print(ruff_format_usage)
```

```python
# ── isort — sort and group imports ────────────────────────────────────────
# isort automatically reorders and groups import statements.
# It's compatible with black and can be configured to match it.

isort_example = '''
# BEFORE isort:
from mypackage import utils
import os
import json
from collections import Counter
import requests
import sys
from pathlib import Path

# AFTER isort:
import json
import os
import sys
from collections import Counter
from pathlib import Path

import requests

from mypackage import utils
'''
print(isort_example)

isort_usage = """
# Install and run
pip install isort
isort .                      # sort all imports in project
isort --check-only .         # check without modifying
isort --diff myfile.py       # show what would change

# Configuration (pyproject.toml):
[tool.isort]
profile = "black"            # compatible with black
line_length = 88
"""
print(isort_usage)
```

```python
# ── autopep8 — conservative auto-fixer ────────────────────────────────────
# autopep8 only fixes PEP 8 violations — it doesn't reformat style.
# Less opinionated than black; makes minimal changes.

autopep8_usage = """
pip install autopep8

# Fix a file in place
autopep8 --in-place myfile.py

# More aggressive fixes
autopep8 --in-place --aggressive --aggressive myfile.py

# Show diff without writing
autopep8 --diff myfile.py
"""
print(autopep8_usage)
```

## 13. Linters

A **linter** analyses code for errors, style violations, and suspicious patterns without running it. Unlike formatters (which fix things), linters **report** what's wrong so you can fix it — or suppress it with a comment.

```python
# ── ruff — the modern all-in-one linter ──────────────────────────────────
# ruff is a Rust-based linter that replaces: flake8, isort, pyupgrade,
# pep8-naming, flake8-bugbear, and many more — all in one tool,
# running 10-100x faster.

ruff_usage = """
pip install ruff

# Check for issues
ruff check .

# Auto-fix safe issues
ruff check --fix .

# Show explanation for a specific rule
ruff rule E501
"""

ruff_config = """
# pyproject.toml
[tool.ruff]
line-length = 88
target-version = "py312"

[tool.ruff.lint]
select = [
    "E",   # pycodestyle errors
    "W",   # pycodestyle warnings
    "F",   # pyflakes (undefined names, unused imports)
    "I",   # isort
    "N",   # pep8-naming
    "UP",  # pyupgrade (modernise syntax)
    "B",   # flake8-bugbear (common bugs)
    "C4",  # flake8-comprehensions
    "SIM", # flake8-simplify
    "TCH", # flake8-type-checking
]
ignore = [
    "E501",  # line too long — handled by formatter
]

[tool.ruff.lint.per-file-ignores]
"tests/*" = ["S101"]   # allow assert in tests
"__init__.py" = ["F401"]  # allow unused imports in __init__
"""
print(ruff_usage)
print(ruff_config)
```

```python
# ── Common ruff/flake8 rule codes ─────────────────────────────────────────
rules = """
E1xx  Indentation
  E101  indentation contains mixed spaces and tabs
  E111  indentation is not a multiple of 4 spaces
  E116  unexpected indentation (comment)

E2xx  Whitespace
  E201  whitespace after '('
  E211  whitespace before '('
  E225  missing whitespace around operator
  E231  missing whitespace after ','
  E251  unexpected spaces around keyword / parameter default
  E261  at least two spaces before inline comment
  E262  inline comment should start with '# '

E3xx  Blank lines
  E301  expected 1 blank line before nested definition
  E302  expected 2 blank lines before function/class
  E303  too many blank lines
  E305  expected 2 blank lines after top-level function/class

E4xx  Imports
  E401  multiple imports on one line
  E402  module import not at top of file

E5xx  Line length
  E501  line too long

E7xx  Statement
  E711  comparison to None (use 'is' or 'is not')
  E712  comparison to True/False (use truthiness)
  E721  do not compare types, use isinstance()
  E741  ambiguous variable name (l, O, I)

W    Warnings
  W291  trailing whitespace
  W292  no newline at end of file
  W605  invalid escape sequence

F    Pyflakes
  F401  module imported but unused
  F403  'from module import *' used
  F811  redefinition of unused name
  F841  local variable assigned but never used

N    Naming (pep8-naming)
  N801  class names should use PascalCase
  N802  function names should be lowercase
  N803  argument names should be lowercase
  N806  variable in function should be lowercase
"""
print(rules)
```

```python
# ── AFTER: fixed ──────────────────────────────────────────────────────────

import re
import sys


def validate_and_score(
    user_input: list,
    weights: list[float],
    normalize: bool = True,
) -> tuple[float, list[str]]:
    """
    Validate a list of values and compute a weighted score.

    Args:
        user_input: List of values to score. None values are skipped.
        weights: Per-element weights. If shorter than user_input,
            remaining elements are given weight 1.0.
        normalize: Divide total score by input length. Defaults to True.

    Returns:
        A (score, errors) tuple where errors is a list of
        human-readable problem descriptions.
    """
    score = 0.0
    errors = []

    for i, item in enumerate(user_input):
        if item is None:
            errors.append(f"item {i} is None")
            continue

        if isinstance(item, str):
            item = item.strip()

        weight = weights[i] if i < len(weights) and weights[i] is not None else 1.0
        score += float(item) * weight

    if normalize:
        score /= len(user_input)

    return score, errors


# Test it
score, errs = validate_and_score(["10", "20", None, "30"], [0.5, 1.0, 1.0, 2.0])
print(f"Score: {score:.2f}")
print(f"Errors: {errs}")

print()
print("VIOLATIONS FIXED:")
fixes = """
  E401  import re,sys            → separate lines
  F403  from math import *       → removed (sqrt not used)
  E302  missing 2 blank lines    → added before function
  N802  ValidateAndScore         → validate_and_score (snake_case)
  N803  Weights                  → weights (lowercase arg)
  E211  func( arg )              → func(arg)
  N806  Score, Errors            → score, errors (lowercase)
  E702  Score=0; Errors=[]       → separate lines
  E402  import json (local)      → removed (not needed)
  B007  range(len(...))          → enumerate()
  E711  item == None             → item is None
  E721  type(item) == str        → isinstance(item, str)
  E711  Weights[i] != None       → weights[i] is not None
  E203  Weights[i] != None :     → no space before colon
  E711  normalize == True        → normalize (truthiness)
  E203  else :                   → else:
"""
print(fixes)
```
