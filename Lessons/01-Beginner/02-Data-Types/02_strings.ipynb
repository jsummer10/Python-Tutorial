## 1. Creating Strings

Python gives you several ways to create a string literal.

```python
# Single and double quotes are interchangeable
s1 = 'Hello'
s2 = "World"

# Use the opposite quote to embed quotes without escaping
s3 = "It's a great day"
s4 = 'She said "hello"'

# Triple quotes — span multiple lines
s5 = """Line one
Line two
Line three"""

# Empty string
s6 = ""

# String constructor
s7 = str(42)       # '42'
s8 = str(3.14)     # '3.14'

for s in [s1, s2, s3, s4, s5, s6, s7, s8]:
    print(repr(s))
```

## 2. Indexing & Slicing

Strings are **sequences** — each character has a numeric index, starting at `0`.

```python
word = "Python"
#        P  y  t  h  o  n
# Index: 0  1  2  3  4  5
# Neg:  -6 -5 -4 -3 -2 -1

print(word[0])   # First character
print(word[-1])  # Last character
print(word[2])   # Third character
```

```python
# Slicing: string[start:stop:step]
s = "Hello, World!"

print(s[0:5])     # 'Hello'   — chars 0 to 4
print(s[7:])      # 'World!'  — from index 7 to end
print(s[:5])      # 'Hello'   — from start to index 4
print(s[-6:])     # 'World!'  — last 6 characters
print(s[::2])     # every 2nd character
print(s[::-1])    # reversed string!
```

```python
# Strings are immutable — you CANNOT change a character in place
try:
    s[0] = 'h'
except TypeError as e:
    print(f"Error: {e}")

# To 'change' a string, create a new one
new_s = 'h' + s[1:]
print(new_s)
```

## 3. String Operators

Python overloads `+`, `*`, and `in` to work with strings.

```python
a = "Hello"
b = "World"

# Concatenation
print(a + ", " + b + "!")   # 'Hello, World!'

# Repetition
print("ha" * 3)              # 'hahaha'
print("-" * 30)              # divider line

# Membership
print("ello" in a)           # True
print("xyz" in a)            # False
print("xyz" not in a)        # True

# Comparison (lexicographic)
print("apple" < "banana")    # True — 'a' comes before 'b'
print("abc" == "abc")        # True
print("abc" == "ABC")        # False — case-sensitive!

# Length
print(len(a))                # 5
```

## 4. Essential String Methods

Python strings have a rich library of built-in methods. They all return **new strings** (remember: strings are immutable).

```python
# Case conversion
s = "hello, WORLD! this is Python."

print(s.upper())       # ALL CAPS
print(s.lower())       # all lowercase
print(s.title())       # Title Case
print(s.capitalize())  # First letter capital, rest lower
print(s.swapcase())    # Flip the case of each letter
```

```python
# Alignment and padding
name = "Alice"

print(name.ljust(20, '.'))   # Left-align in 20 chars
print(name.rjust(20, '.'))   # Right-align in 20 chars
print(name.center(20, '-'))  # Center in 20 chars
print('42'.zfill(6))         # Zero-pad a number string → '000042'
```

```python
# Replacing content
sentence = "The cat sat on the mat"

print(sentence.replace('at', 'og'))          # Replace ALL occurrences
print(sentence.replace('at', 'og', 1))       # Replace only FIRST occurrence

# Expanding tabs
tabbed = "col1\tcol2\tcol3"
print(tabbed.expandtabs(10))  # Replace \t with spaces
```

## 5. Searching & Testing Strings

```python
# Finding substrings
s = "the quick brown fox jumps over the lazy dog"

# find() returns index, or -1 if not found
print(s.find('fox'))        # 16
print(s.find('cat'))        # -1

# index() is like find() but raises ValueError if not found
print(s.index('fox'))       # 16
try:
    s.index('cat')
except ValueError as e:
    print(f"Not found: {e}")

# rfind/rindex — search from the RIGHT
print(s.rfind('the'))       # 31 (last occurrence)

# count — how many times does a substring appear?
print(s.count('the'))       # 2
print(s.count('o'))         # 4
```

```python
# Boolean tests — all return True or False
print("hello".startswith("hel"))   # True
print("hello".endswith("lo"))      # True

# startswith/endswith also accept a tuple of options
filename = "report.pdf"
print(filename.endswith((".pdf", ".docx", ".txt")))  # True

print("-" * 30)

# Character-type tests
print("abc123".isalnum())    # True  — letters and/or digits
print("abc".isalpha())       # True  — only letters
print("123".isdigit())       # True  — only digits
print("   ".isspace())       # True  — only whitespace
print("Hello World".istitle()) # True — title case
print("HELLO".isupper())     # True  — all uppercase
print("hello".islower())     # True  — all lowercase
print("abc123".isidentifier()) # True — valid Python identifier
```

## 6. Splitting, Joining & Stripping

```python
# split() — break a string into a list
csv = "apple,banana,cherry,date"
print(csv.split(','))          # split on comma

sentence = "  one  two  three  "
print(sentence.split())        # no arg: split on any whitespace, strips extras

# Limit the number of splits with maxsplit
print(csv.split(',', maxsplit=2))   # ['apple', 'banana', 'cherry,date']

# rsplit() — split from the right
print(csv.rsplit(',', maxsplit=1))  # ['apple,banana,cherry', 'date']

# splitlines() — split on line breaks
text = "line1\nline2\nline3"
print(text.splitlines())
```

```python
# join() — the OPPOSITE of split()
# It's called on the SEPARATOR, not the list
words = ['apple', 'banana', 'cherry']

print(", ".join(words))    # 'apple, banana, cherry'
print(" | ".join(words))   # 'apple | banana | cherry'
print("".join(words))      # 'applebananacherry'
print("\n".join(words))    # each word on its own line

# Efficient: join is much faster than concatenating in a loop
letters = ['P', 'y', 't', 'h', 'o', 'n']
print("".join(letters))    # 'Python'
```

```python
# Stripping whitespace (or other characters)
s = "   Hello, World!   "

print(repr(s.strip()))     # Remove both ends
print(repr(s.lstrip()))    # Remove left only
print(repr(s.rstrip()))    # Remove right only

# Strip specific characters (set of chars, not a prefix/suffix)
dirty = "***important***"
print(dirty.strip('*'))    # 'important'

# removeprefix / removesuffix (Python 3.9+)
url = "https://example.com"
print(url.removeprefix("https://"))  # 'example.com'

filename = "report.pdf"
print(filename.removesuffix(".pdf")) # 'report'
```

## 7. String Formatting: Three Generations

Python has evolved through three major formatting styles over the years.

```python
name = "Alice"
score = 95.678
rank = 1

# ── Generation 1: % formatting (old-style, C-inspired) ──────────────────────
print("Name: %s, Score: %.2f, Rank: %d" % (name, score, rank))
# Useful format codes: %s = string, %d = int, %f = float, %x = hex, %o = octal

# ── Generation 2: str.format() (Python 2.6+) ────────────────────────────────
print("Name: {}, Score: {:.2f}, Rank: {}".format(name, score, rank))
print("Name: {n}, Score: {s:.2f}".format(n=name, s=score))  # named args

# ── Generation 3: f-strings (Python 3.6+, recommended) ─────────────────────
print(f"Name: {name}, Score: {score:.2f}, Rank: {rank}")
```

## 8. f-Strings — Deep Dive

f-strings (formatted string literals) are the **modern, recommended** way to format strings. They're fast, readable, and powerful.

```python
# Basics: any Python expression can go inside {}
x = 10
y = 3

print(f"x = {x}")
print(f"{x} + {y} = {x + y}")
print(f"x squared = {x ** 2}")
print(f"Is x > y? {x > y}")
print(f"Upper: {'hello'.upper()}")
```

```python
# Format specifications: {value:format_spec}
pi = 3.141592653589793
big = 1_234_567.89

# Floats
print(f"{pi:.2f}")       # 2 decimal places         → 3.14
print(f"{pi:.5f}")       # 5 decimal places         → 3.14159
print(f"{pi:e}")         # scientific notation      → 3.141593e+00
print(f"{pi:.3e}")       # scientific, 3 decimals   → 3.142e+00

# Numbers with separators and width
print(f"{big:,.2f}")     # comma separator          → 1,234,567.89
print(f"{big:_.2f}")     # underscore separator     → 1_234_567.89
print(f"{42:08d}")       # zero-padded to 8 chars   → 00000042
print(f"{42:+d}")        # always show sign         → +42
print(f"{-42:+d}")       #                          → -42
```

```python
# Alignment within a field width
for item, price in [("Apple", 1.5), ("Banana", 0.75), ("Dragonfruit", 4.99)]:
    print(f"{item:<15} ${price:>6.2f}")   # left-align name, right-align price

print()
# < left   > right   ^ center   = padding between sign and digits
print(f"{'left':<20}|")
print(f"{'right':>20}|")
print(f"{'center':^20}|")
print(f"{'fill':*^20}|")
```

```python
# Number bases
n = 255
print(f"{n:d}   decimal")
print(f"{n:b}   binary")
print(f"{n:o}   octal")
print(f"{n:x}   hex (lower)")
print(f"{n:X}   hex (upper)")
print(f"{n:#x}  hex with prefix")
print(f"{n:#b}  binary with prefix")
```

```python
# Percentages and datetime
ratio = 0.8765
print(f"{ratio:.1%}")    # 87.6%

from datetime import datetime
now = datetime(2026, 5, 16, 9, 30)
print(f"{now:%Y-%m-%d}")          # 2026-05-16
print(f"{now:%B %d, %Y}")         # May 16, 2026
print(f"{now:%I:%M %p}")          # 09:30 AM
```

```python
# Python 3.8+: self-documenting expressions with =
# Great for debugging!
x = 42
pi = 3.14159
items = [1, 2, 3]

print(f"{x = }")              # x = 42
print(f"{pi = :.2f}")         # pi = 3.14
print(f"{len(items) = }")     # len(items) = 3
print(f"{x * 2 + 1 = }")     # x * 2 + 1 = 85
```

```python
# Python 3.12+: f-strings can be nested and multi-line
data = {"name": "Alice", "score": 97.5}

# Nested f-string (dynamic format spec)
width = 10
precision = 3
value = 3.14159
print(f"{value:{width}.{precision}f}")  # width and precision from variables

# Multi-expression
language = "Python"
version = 3.14
print(f"{'🐍 ' + language + ' ' + str(version):^30}")
```

## 9. t-Strings — Template Strings (Python 3.14+)

> **Requires Python 3.14+** (released October 2025), introduced via [PEP 750](https://peps.python.org/pep-0750/).

t-strings look just like f-strings but use a `t` prefix. The key difference: **they don't immediately produce a string**. Instead, they produce a `Template` object that preserves the structure — letting you (or a library) safely process interpolations before committing to a final string.

This solves security problems that f-strings can't address, like **SQL injection** and **XSS**.

```python
# ── Check Python version first ─────────────────────────────────────────────
import sys
print(f"Python version: {sys.version}")

if sys.version_info < (3, 14):
    print("⚠️  t-strings require Python 3.14+.")
    print("   Cells below are for illustration — run them on Python 3.14.")
else:
    print("✅ t-strings are available!")
```

```python
# ── Python 3.14+ only ──────────────────────────────────────────────────────
# from string.templatelib import Template  # auto-imported with t-strings

# SYNTAX: t-strings look just like f-strings, but with a t prefix
#
#   f"{name}"  →  produces a str immediately
#   t"{name}"  →  produces a Template object

name = "World"

# f-string: immediate evaluation
f_result = f"Hello, {name}!"
print(type(f_result), repr(f_result))

# t-string: deferred evaluation — this is pseudocode for Python < 3.14
# t_result = t"Hello, {name}!"
# print(type(t_result))   # <class 'string.templatelib.Template'>
# print(t_result.strings) # ('Hello, ', '!')  ← static parts
# print(t_result.interpolations[0].value) # 'World' ← interpolated value
```

```python
# ── Understanding the Template structure ──────────────────────────────────
#
# Given: t"Hello, {name}! Today is {day}."
#
# The Template object contains:
#   .strings        → ('Hello, ', '! Today is ', '.')   ← static text parts
#   .interpolations → (Interpolation(value='Alice', ...),
#                       Interpolation(value='Friday', ...))  ← dynamic parts
#
# This lets a processor DECIDE how to handle each interpolation:
#   - escape HTML entities before inserting into HTML
#   - use parameter binding for SQL (never string-interpolate SQL!)
#   - sanitise before logging

# Simulating what t-strings give you (Python 3.14):
# ┌──────────────────────────────────────────────────────────────┐
# │ name = "Alice"                                               │
# │ day  = "Friday"                                              │
# │ tmpl = t"Hello, {name}! Today is {day}."                    │
# │                                                              │
# │ tmpl.strings          → ('Hello, ', '! Today is ', '.')     │
# │ tmpl.interpolations[0].value  → 'Alice'                     │
# │ tmpl.interpolations[1].value  → 'Friday'                    │
# └──────────────────────────────────────────────────────────────┘

print("Template structure diagram printed above — see comments.")
```

```python
# ── Real-world use case 1: XSS-safe HTML templating ───────────────────────
#
# f-strings are DANGEROUS with user input in HTML context:
#
#   user_input = "<script>alert('pwned')</script>"
#   html = f"<p>{user_input}</p>"   ← injects raw script tag! ⚠️
#
# With t-strings, a library's html() processor can escape first:
#
#   safe_html = html(t"<p>{user_input}</p>")
#   # → "<p>&lt;script&gt;alert('pwned')&lt;/script&gt;</p>"  ✅
#
# Simulation of what an html() processor does:

from html import escape

def simulate_html_processor(static_parts, dynamic_values):
    """Interleave static parts with HTML-escaped dynamic values."""
    result = ""
    for i, part in enumerate(static_parts):
        result += part
        if i < len(dynamic_values):
            result += escape(str(dynamic_values[i]))   # ← safe!
    return result

user_input = "<script>alert('pwned')</script>"
static = ("<p>", "</p>")
dynamic = (user_input,)

unsafe = f"<p>{user_input}</p>"
safe   = simulate_html_processor(static, dynamic)

print(f"Unsafe f-string: {unsafe}")
print(f"Safe t-string:   {safe}")
```

```python
# ── Real-world use case 2: Safe SQL queries ────────────────────────────────
#
# NEVER interpolate user input into SQL strings — SQL injection!
#
#   user_id = "1 OR 1=1"  ← attacker input
#   query = f"SELECT * FROM users WHERE id = {user_id}"  ← VULNERABLE ⚠️
#
# With t-strings, a DB library could extract parameters automatically:
#
#   query, params = sql(t"SELECT * FROM users WHERE id = {user_id}")
#   cursor.execute(query, params)  ← safe parameterized query ✅
#
# Simulation:

def simulate_sql_processor(static_parts, dynamic_values):
    """Build a parameterized SQL query from template parts."""
    query = "?".join(static_parts)  # replace each interpolation with ?
    params = tuple(dynamic_values)
    return query, params

user_id = "1 OR 1=1"   # attacker input
static  = ("SELECT * FROM users WHERE id = ", " AND active = ", "")
dynamic = (user_id, True)

query, params = simulate_sql_processor(static, dynamic)
print(f"Query:  {query}")
print(f"Params: {params}")
print("The DB driver safely binds params — no injection possible ✅")
```

```python
# ── f-strings vs t-strings: side-by-side comparison ───────────────────────

comparison = [
    ("Feature",         "f-string",           "t-string"),
    ("Prefix",          "f or F",              "t or T"),
    ("Result type",     "str",                 "Template object"),
    ("Evaluation",      "Immediate",           "Deferred"),
    ("Security",        "Unsafe w/ user input","Safe — library decides"),
    ("Use case",        "Simple formatting",   "Templating/DSLs/SQL/HTML"),
    ("Python version",  "3.6+",                "3.14+"),
]

col_w = [22, 24, 24]
sep = "+" + "+".join("-" * w for w in col_w) + "+"
print(sep)
for i, row in enumerate(comparison):
    line = "|".join(f" {row[j]:<{col_w[j]-1}}" for j in range(3))
    print("|" + line + "|")
    if i == 0:
        print(sep)
print(sep)
```

## 10. Raw Strings & Escape Sequences

Escape sequences let you embed special characters in strings. Raw strings (`r"..."`) turn escaping off.

```python
# Common escape sequences
print("Tab:\there")           # \t = tab
print("New\nline")            # \n = newline
print("Back\\slash")          # \\ = literal backslash
print("Quote: \"hello\"")    # \" = escaped double quote
print("Bell: \a")             # \a = bell (you may hear it!)
print("Unicode: \u2764")      # \uXXXX = unicode character → ❤
print("Unicode: \U0001F40D")  # \UXXXXXXXX = full unicode  → 🐍
print("Hex: \x41")            # \xHH = hex character       → A
print("Octal: \101")          # \OOO = octal character     → A
```

```python
# Raw strings: backslashes are treated literally (not as escapes)
# Very common for Windows file paths and regular expressions

# Without raw string — need to double every backslash
path1 = "C:\\Users\\Alice\\Documents\\file.txt"

# With raw string — much cleaner
path2 = r"C:\Users\Alice\Documents\file.txt"

print(path1)
print(path2)
print(path1 == path2)  # True — identical strings

# Common in regex
import re
pattern = r"\d{3}-\d{4}"  # match phone numbers like 555-1234
print(re.findall(pattern, "Call 555-1234 or 999-8765 today!"))
```

## 11. Multiline Strings & `textwrap`

```python
# Triple-quoted strings preserve newlines and indentation
poem = """
Roses are red,
Violets are blue,
Python is awesome,
And so are you!
"""
print(poem)

# Implicit string concatenation — adjacent string literals are joined
message = (
    "This is a very long message that would exceed "
    "the recommended line length, so we split it "
    "across multiple lines in the source code."
)
print(message)
```

```python
import textwrap

long_text = (
    "Python is a versatile, high-level programming language known for its "
    "clean syntax and readability. It supports multiple programming paradigms "
    "and has a large standard library."
)

# wrap: returns a list of lines
print(textwrap.wrap(long_text, width=40))
print()

# fill: returns a single wrapped string
print(textwrap.fill(long_text, width=40))
print()

# dedent: strip common leading whitespace (useful with triple-quoted strings)
indented = """
    def hello():
        print('Hello!')
"""
print(textwrap.dedent(indented))

# indent: add a prefix to each line
print(textwrap.indent(long_text[:80], prefix="> "))
```

## 12. Unicode & Encoding

```python
# Python 3 strings are Unicode by default
emoji  = "🐍"
arabic = "مرحبا"
japanese = "こんにちは"
accented = "café"

for s in [emoji, arabic, japanese, accented]:
    print(f"{s!r:20} len={len(s)}")
```

```python
# ord() and chr() — convert between characters and Unicode code points
print(ord('A'))       # 65
print(ord('🐍'))      # 128013
print(chr(65))        # 'A'
print(chr(0x2764))    # '❤'

# Unicode escape forms
print("\u2603")   # ☃ snowman
print("\U0001F600")  # 😀 emoji
print("\N{SNOWMAN}") # ☃ by name
```

```python
# Encoding / decoding between str and bytes
s = "Hello, café! 🐍"

# Encode str → bytes
utf8_bytes = s.encode('utf-8')
print(f"UTF-8:  {utf8_bytes}")
print(f"Length of bytes: {len(utf8_bytes)} (vs {len(s)} chars)")

# Decode bytes → str
decoded = utf8_bytes.decode('utf-8')
print(f"Decoded: {decoded}")

# Error handling during encoding
s2 = "Hello 🐍"
print(s2.encode('ascii', errors='ignore'))    # skip unencodable chars
print(s2.encode('ascii', errors='replace'))   # replace with ?
print(s2.encode('ascii', errors='xmlcharrefreplace'))  # XML ref
```

## 13. Performance Tips

Strings are immutable, so concatenation inside loops can be slow. Here are the best practices.

```python
import time

n = 10_000

# ❌ Slow: += in a loop creates a new string each iteration — O(n²)
start = time.perf_counter()
result = ""
for i in range(n):
    result += str(i)
slow_time = time.perf_counter() - start

# ✅ Fast: collect parts, then join — O(n)
start = time.perf_counter()
parts = []
for i in range(n):
    parts.append(str(i))
result = "".join(parts)
fast_time = time.perf_counter() - start

# ✅ Even faster: use a generator expression with join
start = time.perf_counter()
result = "".join(str(i) for i in range(n))
gen_time = time.perf_counter() - start

print(f"+=  loop:   {slow_time*1000:.3f} ms")
print(f"list+join:  {fast_time*1000:.3f} ms")
print(f"gen+join:   {gen_time*1000:.3f} ms")
```

```python
# String interning — Python caches short strings automatically
# (implementation detail, not something to rely on)
a = "hello"
b = "hello"
print(a is b)       # Likely True — same interned object

c = "hello world"   # Space = not always interned
d = "hello world"
print(c == d)       # True — same VALUE

# Use == to compare string VALUES; use 'is' only for None checks
```

```python
# 🏆 Practice: Build a text report formatter

def format_report(title, data, width=40):
    """
    Format a simple text report using string methods.
    data: list of (label, value) tuples
    """
    lines = []
    lines.append("=" * width)
    lines.append(title.upper().center(width))
    lines.append("=" * width)
    for label, value in data:
        lines.append(f"  {label:<18} {str(value):>{width - 22}}")
    lines.append("-" * width)
    return "\n".join(lines)


report = format_report(
    "Student Report Card",
    [
        ("Name:",        "Alice Johnson"),
        ("Mathematics:", "A+"),
        ("English:",     "A"),
        ("Science:",     "B+"),
        ("History:",     "A-"),
        ("GPA:",         "3.85 / 4.0"),
    ]
)

print(report)
```
