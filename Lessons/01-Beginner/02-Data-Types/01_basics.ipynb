# Introduction to Data Types

In this notebook, you'll learn about Python's core **data types** — the building blocks of every program you'll ever write.

## What You'll Learn
- `int` — Integers (whole numbers)
- `float` — Floating-point numbers (decimals)
- `str` — Strings (text)
- `bool` — Booleans (True/False)
- `list` — Lists (ordered, mutable collections)
- `tuple` — Tuples (ordered, immutable collections)
- `dict` — Dictionaries (key-value pairs)
- `set` — Sets (unique, unordered collections)
- `NoneType` — The special `None` value

---

## 1. Checking Data Types with `type()`

Python has a built-in function called `type()` that tells you what type a value is. We'll use it throughout this tutorial.

```python
print(type(42))         # int
print(type(3.14))       # float
print(type("hello"))    # str
print(type(True))       # bool
print(type(None))       # NoneType
```

---

## 2. Integers (`int`)

Integers are **whole numbers** — positive, negative, or zero. No decimal point.

```python
age = 25
temperature = -7
year = 2026

print(f"Age: {age}, Type: {type(age)}")
print(f"Temperature: {temperature}, Type: {type(temperature)}")
print(f"Year: {year}, Type: {type(year)}")
```

```python
# Integer arithmetic
a = 10
b = 3

print(f"Addition:       {a} + {b} = {a + b}")
print(f"Subtraction:    {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division:       {a} / {b} = {a / b}")    # Returns float!
print(f"Floor Division: {a} // {b} = {a // b}")  # Returns int
print(f"Modulo:         {a} % {b} = {a % b}")    # Remainder
print(f"Exponent:       {a} ** {b} = {a ** b}")
```

---

## 3. Floats (`float`)

Floats are **numbers with a decimal point**. They're used when precision matters.

```python
pi = 3.14159
price = 19.99
negative_float = -0.5

print(f"Pi: {pi}, Type: {type(pi)}")
print(f"Price: {price}, Type: {type(price)}")

# Dividing two ints always gives a float
result = 7 / 2
print(f"7 / 2 = {result}, Type: {type(result)}")
```

```python
# ⚠️ Floating-point precision quirk
# This is a well-known behavior in all programming languages
print(0.1 + 0.2)           # Not exactly 0.3!
print(round(0.1 + 0.2, 2)) # Use round() to fix it
```

---

## 4. Strings (`str`)

Strings are **sequences of characters** — text. Use single `'` or double `"` quotes.

```python
name = "Alice"
greeting = 'Hello, World!'
multi_line = """This is
a multi-line
string."""

print(name)
print(greeting)
print(multi_line)
```

```python
# String operations
first = "Hello"
second = "Python"

# Concatenation
print(first + ", " + second + "!")

# Repetition
print("-" * 20)

# Length
print(f"Length of '{second}': {len(second)}")

# Indexing (0-based)
print(f"First character: {second[0]}")
print(f"Last character:  {second[-1]}")

# Slicing
print(f"First 3 chars: {second[0:3]}")
```

```python
# Useful string methods
sentence = "  Python is Amazing!  "

print(sentence.strip())        # Remove whitespace
print(sentence.lower())        # Lowercase
print(sentence.upper())        # Uppercase
print(sentence.replace("Amazing", "Fun"))  # Replace
print("Python" in sentence)    # Check membership

# f-strings (modern, recommended way to format strings)
name = "Bob"
score = 95.5
print(f"{name} scored {score:.1f}%")
```

---

## 5. Booleans (`bool`)

Booleans have only **two values**: `True` or `False`. They're the foundation of logic and control flow.

```python
is_sunny = True
is_raining = False

print(is_sunny, type(is_sunny))
print(is_raining, type(is_raining))

# Comparison operators return booleans
print(10 > 5)   # True
print(10 == 5)  # False
print(10 != 5)  # True
print(3 <= 3)   # True
```

```python
# Logical operators: and, or, not
has_ticket = True
is_adult = False

print(f"Has ticket AND is adult: {has_ticket and is_adult}")
print(f"Has ticket OR is adult:  {has_ticket or is_adult}")
print(f"NOT has ticket:          {not has_ticket}")

# Truthy and Falsy values
print("\nTruthy/Falsy:")
print(bool(0))      # False — zero is falsy
print(bool(42))     # True  — non-zero is truthy
print(bool(""))     # False — empty string is falsy
print(bool("hi"))   # True  — non-empty string is truthy
print(bool([]))     # False — empty list is falsy
```

## 6. Lists (`list`)

Lists are **ordered, mutable collections**. They can hold any mix of data types.

```python
fruits = ["apple", "banana", "cherry"]
numbers = [1, 2, 3, 4, 5]
mixed = [42, "hello", True, 3.14, None]

print(fruits)
print(f"Type: {type(fruits)}")
print(f"Length: {len(fruits)}")

# Indexing
print(f"First: {fruits[0]}")
print(f"Last:  {fruits[-1]}")

# Slicing
print(f"Slice [0:2]: {fruits[0:2]}")
```

```python
# Modifying lists (mutable!)
fruits = ["apple", "banana", "cherry"]

fruits.append("date")          # Add to end
print(f"After append: {fruits}")

fruits.insert(1, "blueberry")  # Insert at index
print(f"After insert: {fruits}")

fruits.remove("banana")        # Remove by value
print(f"After remove: {fruits}")

popped = fruits.pop()          # Remove and return last
print(f"Popped: {popped}, List: {fruits}")

fruits.sort()                  # Sort in place
print(f"Sorted: {fruits}")
```

## 7. Tuples (`tuple`)

Tuples are like lists, but **immutable** — you can't change them after creation. Great for fixed data.

```python
coordinates = (40.7128, -74.0060)  # New York City lat/lon
rgb_red = (255, 0, 0)
single_item = (42,)  # Note the trailing comma!

print(f"Coordinates: {coordinates}")
print(f"Type: {type(coordinates)}")
print(f"Latitude: {coordinates[0]}")

# Tuple unpacking
lat, lon = coordinates
print(f"Lat: {lat}, Lon: {lon}")
```

```python
# Tuples are immutable — this will raise an error:
try:
    coordinates[0] = 0  # Can't modify!
except TypeError as e:
    print(f"Error: {e}")

# But you can concatenate tuples to make new ones
point_3d = coordinates + (100,)
print(f"3D point: {point_3d}")
```

---

## 8. Dictionaries (`dict`)

Dictionaries store **key-value pairs**. Keys must be unique and immutable; values can be anything.

```python
person = {
    "name": "Alice",
    "age": 30,
    "city": "New York",
    "is_student": False
}

print(person)
print(f"Type: {type(person)}")

# Accessing values
print(f"Name: {person['name']}")
print(f"Age: {person.get('age')}")
print(f"Job (default): {person.get('job', 'Unknown')}")  # Safe with default
```

```python
# Modifying dictionaries
person["age"] = 31           # Update existing key
person["job"] = "Engineer"   # Add new key
del person["is_student"]     # Delete a key

print(person)

# Useful dict methods
print(f"\nKeys:   {list(person.keys())}")
print(f"Values: {list(person.values())}")
print(f"Items:  {list(person.items())}")
```

---

## 9. Sets (`set`)

Sets are **unordered collections of unique elements**. Great for removing duplicates and set math.

```python
colors = {"red", "green", "blue"}
print(colors)
print(f"Type: {type(colors)}")

# Duplicates are automatically removed
numbers = {1, 2, 2, 3, 3, 3, 4}
print(f"Set from duplicates: {numbers}")

# Convert a list to set to remove duplicates
dupes = ["a", "b", "a", "c", "b", "a"]
unique = set(dupes)
print(f"Unique values: {unique}")
```

```python
# Set operations
a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}

print(f"Union (a | b):        {a | b}")        # All elements
print(f"Intersection (a & b): {a & b}")        # Common elements
print(f"Difference (a - b):   {a - b}")        # In a but not b
print(f"Symmetric diff (a^b): {a ^ b}")        # In either but not both
print(f"Is 3 in a? {3 in a}")                  # Membership test
```

## 10. `None` (NoneType)

`None` represents the **absence of a value**. It's Python's equivalent of "nothing" or "null".

```python
result = None
print(result)
print(type(result))

# Always use 'is' to check for None, not '=='
if result is None:
    print("Result has no value yet.")

# Functions without a return statement return None
def greet(name):
    print(f"Hello, {name}!")

output = greet("World")
print(f"Return value: {output}")
```

## 11. Type Conversion

Python can **convert between types** using built-in functions.

```python
# int conversions
print(int(3.9))       # 3   — truncates, doesn't round
print(int("42"))      # 42  — string to int
print(int(True))      # 1
print(int(False))     # 0

# float conversions
print(float(5))       # 5.0
print(float("3.14"))  # 3.14

# str conversions
print(str(100))       # '100'
print(str(3.14))      # '3.14'
print(str(True))      # 'True'

# list/tuple/set conversions
print(list((1, 2, 3)))     # list from tuple
print(tuple([1, 2, 3]))    # tuple from list
print(set([1, 1, 2, 3]))   # set from list
```

```python
# ⚠️ Some conversions will raise errors!
try:
    int("hello")
except ValueError as e:
    print(f"Error: {e}")

try:
    float("abc")
except ValueError as e:
    print(f"Error: {e}")
```

## 12. Checking Types

Python can verify the data type of a variable using isinstance

```python
str_var : str = 'Hello World'
print(f'Is this variable a string? -> {isinstance(str_var, str)}')
```

```python
int_var : int = 1
print(f'Is this variable a string? -> {isinstance(int_var, str)}')
```
