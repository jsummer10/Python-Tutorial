# Iterators and Generators

Welcome to this interactive tutorial! We will explore how Python handles iteration under the hood, and how to create your own iterators and generators to write efficient, pythonic code.

## 1. Iterables vs. Iterators

- **Iterable**: An object you can loop over (e.g., lists, strings, dictionaries). It has an `__iter__()` method.
- **Iterator**: An object representing a stream of data. It has both `__iter__()` and `__next__()` methods.

When you use a `for` loop, Python automatically gets an iterator from the iterable and calls `next()` until it's empty. Let's do this manually.

```python
my_list = ['apple', 'banana', 'cherry']

# Get an iterator from the iterable
my_iterator = iter(my_list)

print(type(my_iterator))

# Fetch items one by one
print(next(my_iterator))  # apple
print(next(my_iterator))  # banana
print(next(my_iterator))  # cherry

# Uncommenting the next line will raise a StopIteration exception
# print(next(my_iterator))
```

## 2. Building a Custom Iterator

To make a custom object iterable, we define a class with `__iter__()` and `__next__()` methods. Let's create a custom counter class.

```python
class MyCounter:
    def __init__(self, start, end):
        self.current = start
        self.end = end

    def __iter__(self):
        # Returns the iterator object itself
        return self

    def __next__(self):
        if self.current > self.end:
            raise StopIteration

        value = self.current
        self.current += 1
        return value

# Using our custom iterator
counter = MyCounter(1, 3)
for num in counter:
    print(num)
```

## 3. Generators: Elegant Iterators

Writing full classes for iterators is verbose. **Generators** are a simpler way to create iterators using functions and the `yield` keyword. 

Every time `yield` is called, the function's execution is paused, its state is saved, and a value is returned to the caller.

```python
def generator_counter(start, end):
    current = start
    while current <= end:
        yield current
        current += 1

gen = generator_counter(1, 3)
print("Object type:", type(gen))

for num in gen:
    print(num)
```

## 4. Generator Expressions & Memory Efficiency

Generator expressions look exactly like list comprehensions, but use parentheses `()` instead of brackets `[]`. 

Because they evaluate **lazily** (generating one item on the fly rather than storing everything in memory at once), they are incredibly memory efficient for large datasets.

```python
import sys

# List comprehension (builds the entire list in memory)
my_list_comp = [x**2 for x in range(10000)]

# Generator expression (generates items one at a time)
my_gen_expr = (x**2 for x in range(10000))

print(f"List size in memory: {sys.getsizeof(my_list_comp):,} bytes")
print(f"Generator size in memory: {sys.getsizeof(my_gen_expr):,} bytes")
```
