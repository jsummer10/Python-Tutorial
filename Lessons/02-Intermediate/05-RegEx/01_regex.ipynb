# Regular Expressions (Regex)

Regular expressions are powerful tools for matching patterns in text. Whether you need to validate user input, extract specific data from a large document, or replace strings, regex is the tool for the job.

In Python, we use the built-in `re` module to work with regular expressions. **Note:** We almost always use 'raw strings' (prefixing the string with `r`, like `r"pattern"`) for regex so that Python doesn't try to interpret special escape characters like `\n`.

## 1. Searching for Patterns: `search` vs `match`

- `re.search()`: Scans the *entire* string and returns the first successful match.
- `re.match()`: Only checks for a match at the *very beginning* of the string.

```python
import re

text = "The quick brown fox jumps over the lazy dog"

# Search looks anywhere in the string
search_result = re.search(r"fox", text)
print("Search for 'fox':", search_result.group() if search_result else "No match")

# Match only looks at the beginning
match_result = re.match(r"fox", text)
print("Match for 'fox':", match_result.group() if match_result else "No match")

# Match works if the pattern is at the start
match_start = re.match(r"The", text)
print("Match for 'The':", match_start.group() if match_start else "No match")
```

## 2. Character Classes and Quantifiers

Regex uses special characters to represent types of characters and how many times they appear:

- `\d`: Any digit (0-9)
- `\w`: Any word character (a-z, A-Z, 0-9, _)
- `\s`: Any whitespace character (space, tab, newline)
- `+`: One or more times
- `*`: Zero or more times
- `?`: Zero or one time (optional)
- `{n}`: Exactly *n* times

```python
invoice_data = "Invoice #INV-59321 is due on 2023-10-15."

# Extract the invoice number (starts with INV-, followed by 5 digits)
invoice_pattern = r"INV-\d{5}"
inv_match = re.search(invoice_pattern, invoice_data)
print("Invoice Number:", inv_match.group())

# Extract the date (4 digits - 2 digits - 2 digits)
date_pattern = r"\d{4}-\d{2}-\d{2}"
date_match = re.search(date_pattern, invoice_data)
print("Due Date:", date_match.group())
```

## 3. Finding All Matches and Substituting

- `re.findall()`: Returns a list of all non-overlapping matches in the string.
- `re.sub()`: Replaces matches with a new string.

```python
contact_info = "Contact Alice at alice@example.com or Bob at bob.smith@work-domain.org."

# 1. Find all email addresses
# Pattern breakdown: word chars/dots/dashes + @ + word chars/dots/dashes + . + 2 to 4 letters
email_pattern = r"[\w.-]+@[\w.-]+\.[a-zA-Z]{2,4}"
emails = re.findall(email_pattern, contact_info)
print("Extracted Emails:", emails)

# 2. Redact email addresses for privacy
redacted_text = re.sub(email_pattern, "[REDACTED]", contact_info)
print("\nRedacted text:", redacted_text)
```

## 4. Groups and Capturing

Parentheses `()` in regex allow you to group parts of your pattern together. This lets you extract specific sub-sections of a matched string.

```python
log_entry = "ERROR 404: Page not found at /users/profile"

# We want to extract the severity (ERROR) and the code (404) separately.
# We wrap the parts we want to extract in parentheses.
log_pattern = r"([A-Z]+)\s(\d{3}):"

match = re.search(log_pattern, log_entry)

if match:
    print("Full match:", match.group(0))  # The entire matched string
    print("Group 1 (Severity):", match.group(1))
    print("Group 2 (Code):", match.group(2))
```
